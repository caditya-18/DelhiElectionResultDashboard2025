# FetchingDataForSingleConstituency

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

# Set up Selenium WebDriver
driver = webdriver.Chrome()  # Make sure you have chromedriver installed
driver.get("https://results.eci.gov.in/ResultAcGenFeb2025/ConstituencywiseU054.htm")

# Wait for the page to load
driver.implicitly_wait(10)

# Find table rows
rows = driver.find_elements(By.XPATH, "//table/tbody/tr")
data = []
for row in rows:
    cols = row.find_elements(By.TAG_NAME, "td")
    cols = [col.text.strip() for col in cols]
    if cols:
        data.append(cols)

# Convert to DataFrame
df = pd.DataFrame(data, columns=["S.N.", "Candidate", "Party", "EVM Votes", "Postal Votes", "Total Votes", "% of Votes"])
# print(df)

# Remove the "% of Votes" column
df = df.drop(columns=["% of Votes"])

# Print and save the modified DataFrame
print(df)
df.to_csv("election_results.csv", index=False)


# Close the browser
driver.quit()
```

# fetchingAllTheDataForAllConstituencies

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Listofconstituency 
constituency_numbers = list(range(1, 71))

# BaseURL
base_url = "https://results.eci.gov.in/ResultAcGenFeb2025/ConstituencywiseU05{}.htm"

# Set up Selenium WebDriver
driver = webdriver.Chrome()  #chromedriver is installed

#emptyListToStoreConstituencyData
all_data = []

# LoopThroughEachConstituency
for num in constituency_numbers:
    url = base_url.format(num)
    driver.get(url)
    
    # Wait for the page to load
    time.sleep(3)
    
    # Find table rows
    rows = driver.find_elements(By.XPATH, "//table/tbody/tr")
    
    for row in rows:
        cols = row.find_elements(By.TAG_NAME, "td")
        cols = [col.text.strip() for col in cols]
        if cols:
            all_data.append([num] + cols)  
    
# ConvertToDataFrame
df = pd.DataFrame(all_data, columns=["Constituency No", "S.N.", "Candidate", "Party", "EVM Votes", "Postal Votes", "Total Votes", "% of Votes"])

# RemoveThe"% of Votes"Column
df = df.drop(columns=["% of Votes"])

# PrintAndSaveTheModifiedDataFrame
print(df)
path = "D:\DelhiElection2025\delhi_election_results.csv"
df.to_csv(path, index=False)

# CloseTheBrowser
driver.quit()
```
