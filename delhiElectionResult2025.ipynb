# FetchingDataForSingleConstituency

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

# Set up Selenium WebDriver
driver = webdriver.Chrome()  # Make sure you have chromedriver installed
driver.get("https://results.eci.gov.in/ResultAcGenFeb2025/ConstituencywiseU054.htm")

# Wait for the page to load
driver.implicitly_wait(10)

# Find table rows
rows = driver.find_elements(By.XPATH, "//table/tbody/tr")
data = []
for row in rows:
    cols = row.find_elements(By.TAG_NAME, "td")
    cols = [col.text.strip() for col in cols]
    if cols:
        data.append(cols)
        
# print(data)

# Convert to DataFrame
df = pd.DataFrame(data, columns=["S.N.", "Candidate", "Party", "EVM Votes", "Postal Votes", "Total Votes", "% of Votes"])
# print(df)

# Remove the "% of Votes" column
df = df.drop(columns=["% of Votes"])

# Print and save the modified DataFrame
print(df)
df.to_csv("election_results.csv", index=False)


# Close the browser
driver.quit()
```

# fetchingAllTheDataForAllConstituencies

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Listofconstituency 
constituency_numbers = list(range(1, 71))

# BaseURL
base_url = "https://results.eci.gov.in/ResultAcGenFeb2025/ConstituencywiseU05{}.htm"

# Set up Selenium WebDriver
driver = webdriver.Chrome()  #chromedriver is installed

#emptyListToStoreConstituencyData
all_data = []

# LoopThroughEachConstituency
for num in constituency_numbers:
    url = base_url.format(num)
    driver.get(url)
    
    # Wait for the page to load
    time.sleep(3)
    
    # Find table rows
    rows = driver.find_elements(By.XPATH, "//table/tbody/tr")
    
    for row in rows:
        cols = row.find_elements(By.TAG_NAME, "td")
        cols = [col.text.strip() for col in cols]
        if cols:
            all_data.append([num] + cols)  
    
# ConvertToDataFrame
df = pd.DataFrame(all_data, columns=["Constituency No", "S.N.", "Candidate", "Party", "EVM Votes", "Postal Votes", "Total Votes", "% of Votes"])

# RemoveThe"% of Votes"Column
df = df.drop(columns=["% of Votes"])

# PrintAndSaveTheModifiedDataFrame
print(df)
path = "D:\DelhiElection2025\delhi_election_results.csv"
df.to_csv(path, index=False)

# CloseTheBrowser
driver.quit()
```

```python
import pdfplumber
import pandas as pd

pdf_path = r"D:\DelhiElection2025\PC_wise_Electors_With_AC.pdf"

data = []
current_pc_no = None  # Store the last valid PC No.
current_pc_name = None  # Store the last valid PC Name

with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        table = page.extract_table()
        if table:
            for row in table:
                # Skip unwanted header rows
                if row[0] == "PC No.":
                    continue
                if "Sum" in str(row[1]):  # Skip summary rows
                    continue

                # Handle empty PC No. and PC Name cases (inherit from previous row)
                if row[0] and row[1]:  
                    current_pc_no = row[0]
                    current_pc_name = row[1]

                elif not row[0] and not row[1]:  
                    row[0] = current_pc_no
                    row[1] = current_pc_name

                # Ensure row has enough columns
                if len(row) >= 8:
                    data.append(row[:8])

# Convert to DataFrame
df = pd.DataFrame(data, columns=["PC No", "PC Name", "ac_no", "ac_name_EN", "Male", "Female", "Third Gender", "Total"])

# Remove unwanted header rows that might have been captured again
df = df[df["PC No"].str.isnumeric()]

# Convert numeric columns
df["Male"] = pd.to_numeric(df["Male"], errors="coerce")
df["Female"] = pd.to_numeric(df["Female"], errors="coerce")
df["Third Gender"] = pd.to_numeric(df["Third Gender"], errors="coerce")
df["Total"] = pd.to_numeric(df["Total"], errors="coerce")

print(df.head())  

# saving this dataframe into csv
csv_path = r"D:\DelhiElection2025\PC_wise_Electors.csv"
df.to_csv(csv_path, index = False, encoding = "utf-8")
```
